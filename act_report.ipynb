# Data Visualization process explained

## Aim:
1. To analyse the relation between retweet and favorite counts with the rating_numerator column.
2. The type of dog with highest average retweet counts and favorite counts.

Kindly note that the master csv file generated after the cleaning process is then used for the analysis process here. The codes for the visualization is 
present in wrangle_act.ipynb


Pre requisites for the analysis:

1.Some columns which do not carry any significance in the analysis further 
is dropped.

2.The dataframe is queried to obtain only those rows where the dog_type value
is not null.

3. The data rows are reduced to include those where the rating_numerator column values are less than 15 to analyse the relation better.

### To analyse the relation between retweet and favorite counts with the rating_numerator column:
    
1. A scatter plot is plotted between the rating_numerator and retweet 
 count column.

As we can see, a relation is established between retweet count 
and ratings column. Higher the ratings, higher is the retweet count values.

2. A scatter plot is plotted between the rating_numerator and favorite 
 count column.

As we can see, a relation is established between favorite count 
and ratings column. Higher the ratings, higher is the favorite count values.

### The type of dog with highest average retweet counts and favorite counts:
1. The dataframe containing those rows which do not have null values in
dog_type column are queried to obtain the resulting dataframe having the
various dog type values.
2. A bar plot is plotted between the dog_type and the retweet count column 
and favorite counts column.

Puppy (Puppo or Pupper) is the dog_type with the most average number of 
favorite counts.

Doggo(normal Dog) and Puppy (Puppo) has the most average retweet counts.

### Observations/Insights:
    1. A relation is established between the rating_numerator and 
    retweet count column. The same goes with rating_numerator and favorite 
    count column too.
    
    2. Puppy (Puppo or Pupper) is the dog_type with the most number of
    favorite counts.
    
    3. Doggo(normal Dog) and Puppy (Puppo) has the most average retweet counts.

### Limitations:
    1. Since the data collected is of small size, the analysis cannot be
    generalised.
    2. The predictions made are based on the analysis, hence are subjected 
    to correction or change.
