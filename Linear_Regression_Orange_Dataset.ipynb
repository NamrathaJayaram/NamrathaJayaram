```python
from sklearn.linear_model import LinearRegression
import numpy as np
model = LinearRegression()
```

```python
X = df["circumference"].values.reshape(-1,1)
y = df["age"]
```

```python
model.fit(X, y)
```

```python
model.score(X, y) #R2 score
```

```python
y_pred = model.predict(X)
```

```python
y_pred
```

```python
y_true = y
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_true, y_pred)
```

```python
print(X,y_true,y_pred, sep = '\n')
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
pol_reg = LinearRegression()
pol_reg.fit(X_poly, y)
y_poly_pred = model.predict(X_poly)


# Visualizing the Polymonial Regression results
def viz_polymonial():
    plt.scatter(X, y, color='red')
    plt.plot(X, y_poly_pred, color='blue')
    plt.xlabel('Circumference')
    plt.ylabel('Age')
    plt.show()
    return
viz_polymonial()
```

```python
mean_squared_error(y_true, y_poly_pred) # observe that the mean squared error has reduced after we increased the model complexity by introducing the ploynomial features od 2nd degree
```
